```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
import os
```

```python
def get_warehouse(url):
    DOWNLOAD_DIR = os.path.expanduser("~/Downloads")
    driver = webdriver.Chrome()
    driver.get(url)
    try:
        button = driver.find_element(By.CSS_SELECTOR, "button[data-name='Button']")
        button.click()
        time.sleep(10)
        files = os.listdir(DOWNLOAD_DIR)
        excel_files = [f for f in files if f.endswith("offers.xlsx")]
        if excel_files:
            excel_file_path = os.path.join(DOWNLOAD_DIR, excel_files[0])
            df_table = pd.read_excel(excel_file_path)
            driver.quit()
            return df_table
    except Exception as e:
        driver.quit()
        print(type(e))
        print(e)
df_table = get_warehouse('https://www.cian.ru/snyat-sklad/')
df_table
```
